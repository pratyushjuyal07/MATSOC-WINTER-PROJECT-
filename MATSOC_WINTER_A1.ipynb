```python
#q2
import networkx as nx
import matplotlib.pyplot as plt

G=nx.Graph()

nodes=[x for x in 'ABCDE']
G.add_nodes_from(nodes)

edges=[('A','B',4) , ('A','C',8), ('A','E',2),
        ('B','D',1), ('B','E',3) ,
        ('C','D',3),
        ('D','E',6) ]
G.add_weighted_edges_from(edges)

nx.draw(G,with_labels=True,node_color='gold', width=[x[2] for x in edges])

d_centrality=nx.degree_centrality(G)
print('Degree Centrality',d_centrality)

wd_centrality={}
for i in nodes:
  eachnodeweight=0
  for j in edges:
    if i in j:
      eachnodeweight+=j[2]
  wd_centrality[i]=eachnodeweight
print('Weighted Deg Centr',wd_centrality)
```

```python
#q3
import networkx as nx

G=nx.DiGraph()

G.add_nodes_from(range(1,6))
G.add_edges_from([ (1,2),(2,3),(3,4),(4,5),(5,4),(2,5),(1,3),(4,2) ])

nx.draw(G,with_labels=True,node_color='lightpink',arrowstyle='-|>')

print('(Node,InDegree )',G.in_degree)
print('(Node,OutDegree)',G.out_degree)
```
